```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time = np.array([i for i in range(10)])
```

```python
setpoints = np.array([30 for i in range(10)])
setpoints[0:4] = 0
```

```python
setpoints
```

```python
current_values = np.array([10,10,10,10,15,20,25,30,35,35])
```

```python
plt.plot(time,setpoints)
plt.plot(time,current_values)
plt.xlabel("Time in minutes")
plt.ylabel("Moisture level in %")
plt.title(" Setpoints and current values of moisture sensors")
plt.legend(("Setpoints","Current values"))
```
